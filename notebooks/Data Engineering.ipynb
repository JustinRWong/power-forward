```python
built = pd.DataFrame()
for addy in combined['mapped_addy']:
    retained = f[f['mapped_addy']==addy]
    built = pd.concat([built,retained])
```

```python
station_cols = ['connection_type', 'power_kwh', 'power_kwh_num', 
              'mapped_addy', 'latitude', 'longitude',
               'available_denominator']
time_cols = ['day_of_week', 'day_of_week_num', 'month', 'date', 'hour', 'minutes',
       'seconds',]

## average all columns, while getting rid of time related columns
# drop time related columns since one station should have a single score regardless of time
# ignore charging stations that have no denominator
UR_by_stations_first_group = df.drop(time_cols, axis=1)[df['available_denominator'] != 0].groupby(by=station_cols).agg('mean').reset_index()

address_cols = ['mapped_addy', 'latitude', 'longitude']
UR_by_stations = UR_by_stations_first_group.groupby(address_cols).agg('sum').reset_index()
UR_by_stations.head()

UR_by_stations['availability_rate'] = UR_by_stations['available_numerator']/UR_by_stations['available_denominator']

UR_by_stations['utilization_rate'] = 1-UR_by_stations['availability_rate']

UR_by_stations['total_output_kwh'] = UR_by_stations['power_kwh_num']*UR_by_stations['utilization_rate']

UR_by_stations['utilization_numerator'] = UR_by_stations['available_denominator']*UR_by_stations['utilization_rate']
```

```python
kept_cols = ['mapped_addy', 'latitude', 'longitude', 'power_kwh_num', 'available_denominator', 'available_numerator', 'availability_rate', 'utilization_rate', 'total_output_kwh', 'utilization_numerator']
UR_by_stations[kept_cols].to_csv('utilization_rates_data/final_113.csv')
```
