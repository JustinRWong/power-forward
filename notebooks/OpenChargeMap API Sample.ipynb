## Open Charge Map API Usage

See https://openchargemap.org/site/develop/api

```python
## Define api parameters
maxresults = "10"
key = "10f52d91-4f2c-44d4-aacc-16dff6c88d99"
long = "-122.431297" ## This is san francisco
lat = "37.773972"
distance = "20"
dist_units = "Miles"
example = "https://api.openchargemap.io/v3/poi/?output=json&countrycode=US&maxresults={maxresults}&compact=true&verbose=false&key={key}&longitude={long}&latitude={lat}&distance={distance}&distanceunit={dist_units}".format(key=key, maxresults=maxresults, long=long, lat=lat, distance=distance, dist_units=dist_units)

response = requests.get(example)
json_response = response.json()
```

The `AddressInfo` and `Connections` columns have embedded json. This needs to be flattened.

```python
json_unprocessed_df = pd.DataFrame(json_response)

info(json_unprocessed_df)
```

```python
## Reference: https://towardsdatascience.com/flattening-json-objects-in-python-f5343c794b10
def flatten_json(y):
    out = {}
    
    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
flattend_json_response = [flatten_json(charger) for charger in json_response]
# flattened_dict = flattend_json_response
flattened_json_df = pd.DataFrame(flattend_json_response)

info(flattened_json_df)
```
